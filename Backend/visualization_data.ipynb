```python
def diagnosisAge(dia,mi,ma):
    count = 0
    for i in range(1, 70):
        if dia in data[str(i)]['Diagnosis']:
            if data[str(i)]['Age'] is not None:           
                if data[str(i)]['Age']>=mi and data[str(i)]['Age']<ma:
                    count +=1
    return count
        
```

```python
def sexCount():
    count_m=0
    count_f=0
    for i in range(1, 70):
    
        if data[str(i)]['Sex'] is not None:           
            if data[str(i)]['Sex']=="M":
                count_m +=1
            elif data[str(i)]['Sex']=="F":
                count_f +=1
    return count_m, count_f
```
